```python
import requests
import json
import pymysql
import time
import os
from datetime import datetime
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("COC_API_KEY")
headers = {"Authorization": f"Bearer {API_KEY}"}

db = pymysql.connect(
    host=os.getenv("DB_HOST", "localhost"),
    user=os.getenv("DB_USER", "root"),
    password=os.getenv("DB_PASSWORD"),
    database=os.getenv("DB_NAME", "coc_analytics")
)

cursor = db.cursor()
print("Connected to MySQL successfully!")
print("API headers ready!")
```

```python
BASE_URL = "https://api.clashofclans.com/v1"

def get_player(player_tag):
    tag = player_tag.replace("#", "%23")
    url = f"{BASE_URL}/players/{tag}"
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error fetching player {player_tag}: {response.status_code}")
        return None

def get_clan(clan_tag):
    tag = clan_tag.replace("#", "%23")
    url = f"{BASE_URL}/clans/{tag}"
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error fetching clan {clan_tag}: {response.status_code}")
        return None

def get_warlog(clan_tag):
    tag = clan_tag.replace("#", "%23")
    url = f"{BASE_URL}/clans/{tag}/warlog?limit=20"
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.json().get('items', [])
    else:
        print(f"Error fetching warlog {clan_tag}: {response.status_code}")
        return []

def get_current_war(clan_tag):
    tag = clan_tag.replace("#", "%23")
    url = f"{BASE_URL}/clans/{tag}/currentwar"
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        data = response.json()
        if data.get('state') != 'notInWar':
            return data
    return None

def get_top_clans(location_id, limit=200):
    url = f"{BASE_URL}/locations/{location_id}/rankings/clans?limit={limit}"
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.json().get('items', [])
    else:
        print(f"Error fetching top clans for location {location_id}: {response.status_code}")
        return []

print("Helper functions ready!")
```

```python
def insert_player(player):
    try:
        sql = """
        INSERT IGNORE INTO players 
        (player_tag, name, town_hall_level, exp_level, trophies, best_trophies,
        war_stars, attack_wins, defense_wins, donations, donations_received,
        war_preference, league_name, legend_trophies)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
        """
        values = (
            player.get('tag'),
            player.get('name'),
            player.get('townHallLevel'),
            player.get('expLevel'),
            player.get('trophies'),
            player.get('bestTrophies'),
            player.get('warStars'),
            player.get('attackWins'),
            player.get('defenseWins'),
            player.get('donations'),
            player.get('donationsReceived'),
            player.get('warPreference'),
            player.get('league', {}).get('name'),
            player.get('legendStatistics', {}).get('legendTrophies')
        )
        cursor.execute(sql, values)
        db.commit()
    except Exception as e:
        print(f"Error inserting player {player.get('tag')}: {e}")

def insert_clan(clan, country=None):
    try:
        sql = """
        INSERT IGNORE INTO clans
        (clan_tag, name, clan_level, clan_points, war_wins, war_losses,
        war_ties, war_frequency, war_league, members, country)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
        """
        values = (
            clan.get('tag'),
            clan.get('name'),
            clan.get('clanLevel'),
            clan.get('clanPoints'),
            clan.get('warWins'),
            clan.get('warLosses', 0),
            clan.get('warTies', 0),
            clan.get('warFrequency'),
            clan.get('warLeague', {}).get('name'),
            clan.get('members'),
            country
        )
        cursor.execute(sql, values)
        db.commit()
    except Exception as e:
        print(f"Error inserting clan {clan.get('tag')}: {e}")

def insert_clan_member(player_tag, clan_tag, member):
    try:
        sql = """
        INSERT INTO clan_members
        (player_tag, clan_tag, role, clan_rank, previous_clan_rank)
        VALUES (%s, %s, %s, %s, %s)
        """
        values = (
            player_tag,
            clan_tag,
            member.get('role'),
            member.get('clanRank'),
            member.get('previousClanRank')
        )
        cursor.execute(sql, values)
        db.commit()
    except Exception as e:
        print(f"Error inserting clan member {player_tag}: {e}")

def insert_troop_levels(player_tag, troops):
    try:
        sql = """
        INSERT INTO troop_levels
        (player_tag, troop_name, troop_type, level, max_level, village)
        VALUES (%s, %s, %s, %s, %s, %s)
        """
        values = [
            (
                player_tag,
                troop.get('name'),
                'troop',
                troop.get('level'),
                troop.get('maxLevel'),
                troop.get('village')
            )
            for troop in troops
        ]
        cursor.executemany(sql, values)
        db.commit()
    except Exception as e:
        print(f"Error inserting troops for {player_tag}: {e}")

def insert_spell_levels(player_tag, spells):
    try:
        sql = """
        INSERT INTO troop_levels
        (player_tag, troop_name, troop_type, level, max_level, village)
        VALUES (%s, %s, %s, %s, %s, %s)
        """
        values = [
            (
                player_tag,
                spell.get('name'),
                'spell',
                spell.get('level'),
                spell.get('maxLevel'),
                spell.get('village')
            )
            for spell in spells
        ]
        cursor.executemany(sql, values)
        db.commit()
    except Exception as e:
        print(f"Error inserting spells for {player_tag}: {e}")

def insert_hero_levels(player_tag, heroes):
    try:
        sql = """
        INSERT INTO hero_levels
        (player_tag, hero_name, level, max_level, village)
        VALUES (%s, %s, %s, %s, %s)
        """
        values = [
            (
                player_tag,
                hero.get('name'),
                hero.get('level'),
                hero.get('maxLevel'),
                hero.get('village')
            )
            for hero in heroes
        ]
        cursor.executemany(sql, values)
        db.commit()
    except Exception as e:
        print(f"Error inserting heroes for {player_tag}: {e}")

def insert_war(clan_tag, war):
    try:
        end_time = war.get('endTime', '')
        war_id = f"{clan_tag}_{end_time}"
        sql = """
        INSERT IGNORE INTO wars
        (war_id, clan_tag, opponent_tag, opponent_name, result,
        team_size, clan_stars, opponent_stars, clan_destruction, opponent_destruction, end_time)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
        """
        end_time_formatted = datetime.strptime(end_time[:15], "%Y%m%dT%H%M%S") if end_time else None
        values = (
            war_id,
            clan_tag,
            war.get('opponent', {}).get('tag'),
            war.get('opponent', {}).get('name'),
            war.get('result'),
            war.get('teamSize'),
            war.get('clan', {}).get('stars'),
            war.get('opponent', {}).get('stars'),
            war.get('clan', {}).get('destructionPercentage'),
            war.get('opponent', {}).get('destructionPercentage'),
            end_time_formatted
        )
        cursor.execute(sql, values)
        db.commit()
        return war_id
    except Exception as e:
        print(f"Error inserting war: {e}")
        return None

def insert_war_attacks(war_id, attacks, attacker_tag):
    try:
        sql = """
        INSERT IGNORE INTO war_attacks
        (war_id, attacker_tag, defender_tag, stars, destruction_pct, duration_seconds, attack_order)
        VALUES (%s, %s, %s, %s, %s, %s, %s)
        """
        values = [
            (
                war_id,
                attack.get('attackerTag'),
                attack.get('defenderTag'),
                attack.get('stars'),
                attack.get('destructionPercentage'),
                attack.get('duration'),
                attack.get('order')
            )
            for attack in attacks
        ]
        cursor.executemany(sql, values)
        db.commit()
    except Exception as e:
        print(f"Error inserting war attacks: {e}")

print("All insert functions ready!")
```

```python
# Get clan tags already in database so we can skip them
cursor.execute("SELECT clan_tag FROM clans")
already_fetched = set(row[0] for row in cursor.fetchall())
print(f"Already have {len(already_fetched)} clans in database — will skip these")

total_players = 0
total_clans = 0
total_wars = 0

for country_name, location_id in LOCATIONS.items():
    print(f"\n{'='*50}")
    print(f"Fetching top clans for: {country_name}")
    print(f"{'='*50}")
    
    top_clans = get_top_clans(location_id, limit=200)
    print(f"Found {len(top_clans)} clans")
    
    for i, clan_summary in enumerate(top_clans):
        clan_tag = clan_summary.get('tag')
        
        # Skip if already fetched
        if clan_tag in already_fetched:
            print(f"  Skipping {clan_tag} — already in database")
            continue
        
        clan_data = get_clan(clan_tag)
        if not clan_data:
            continue
        
        insert_clan(clan_data, country=country_name)
        total_clans += 1
        
        members = clan_data.get('memberList', [])
        for member in members:
            player_tag = member.get('tag')
            player_data = get_player(player_tag)
            if not player_data:
                continue
            
            insert_player(player_data)
            insert_clan_member(player_tag, clan_tag, member)
            insert_troop_levels(player_tag, player_data.get('troops', []))
            insert_spell_levels(player_tag, player_data.get('spells', []))
            insert_hero_levels(player_tag, player_data.get('heroes', []))
            
            total_players += 1
            time.sleep(0.05)
        
        warlog = get_warlog(clan_tag)
        for war in warlog:
            war_id = insert_war(clan_tag, war)
            total_wars += 1
        
        if (i + 1) % 10 == 0:
            print(f"  Processed {i+1}/{len(top_clans)} clans | Players: {total_players} | Wars: {total_wars}")
        
        time.sleep(0.1)

print(f"\n{'='*50}")
print(f"SESSION COMPLETE!")
print(f"Clans this session:   {total_clans}")
print(f"Players this session: {total_players}")
print(f"Wars this session:    {total_wars}")
print(f"{'='*50}")
```

```python
# Fetch war attacks from current war for all clans
cursor.execute("SELECT clan_tag FROM clans")
all_clan_tags = [row[0] for row in cursor.fetchall()]

print(f"Fetching current war data for {len(all_clan_tags)} clans...")

total_attacks = 0
wars_found = 0

for i, clan_tag in enumerate(all_clan_tags):
    war_data = get_current_war(clan_tag)
    
    if not war_data:
        continue
    
    # Build war_id same way as before
    end_time = war_data.get('endTime', '')
    war_id = f"{clan_tag}_{end_time}"
    
    # Insert the war itself first
    insert_war(clan_tag, war_data)
    wars_found += 1
    
    # Now extract attacks from each member
    members = war_data.get('clan', {}).get('members', [])
    for member in members:
        attacks = member.get('attacks', [])
        if attacks:
            insert_war_attacks(war_id, attacks, member.get('tag'))
            total_attacks += len(attacks)
    
    # Progress every 50 clans
    if (i + 1) % 50 == 0:
        print(f"  Processed {i+1}/{len(all_clan_tags)} clans | Attacks found: {total_attacks}")
    
    time.sleep(0.1)

print(f"\nDone!")
print(f"Wars with attack data: {wars_found}")
print(f"Total attacks inserted: {total_attacks}")
```

## War Attack Data

The warlog endpoint only gives win/loss/tie results — no individual attack details. For stars, destruction %, and duration I had to hit `/currentwar` separately for each clan. That's why this is a second loop instead of part of the main collection.

```python
tables = ['players', 'clans', 'wars', 'clan_members', 'troop_levels', 'hero_levels', 'war_attacks']

print("="*40)
print("FINAL DATABASE SUMMARY")
print("="*40)
for table in tables:
    cursor.execute(f"SELECT COUNT(*) FROM {table}")
    count = cursor.fetchone()[0]
    print(f"{table:<20} {count:>10,} rows")
print("="*40)
```

```python
# Re-fetch war attacks - only for valid wars now in our database
cursor.execute("SELECT clan_tag FROM clans")
all_clan_tags = [row[0] for row in cursor.fetchall()]

print(f"Re-fetching war attacks for {len(all_clan_tags)} clans...")

total_attacks = 0
wars_found = 0

for i, clan_tag in enumerate(all_clan_tags):
    war_data = get_current_war(clan_tag)
    
    if not war_data:
        continue
    
    state = war_data.get('state')
    if state not in ['inWar', 'warEnded']:
        continue
        
    end_time = war_data.get('endTime', '')
    war_id = f"{clan_tag}_{end_time}"
    
    # Only insert attacks if war exists and is clean
    cursor.execute("SELECT war_id FROM wars WHERE war_id = %s", (war_id,))
    existing_war = cursor.fetchone()
    
    if not existing_war:
        # Insert war first if not exists
        insert_war(clan_tag, war_data)
    
    # Extract and insert attacks
    members = war_data.get('clan', {}).get('members', [])
    for member in members:
        attacks = member.get('attacks', [])
        if attacks:
            insert_war_attacks(war_id, attacks, member.get('tag'))
            total_attacks += len(attacks)
    
    wars_found += 1
    
    if (i + 1) % 50 == 0:
        print(f"  Processed {i+1}/{len(all_clan_tags)} clans | Attacks: {total_attacks}")
    
    time.sleep(0.1)

print(f"\nDone!")
print(f"Wars processed: {wars_found}")
print(f"Total attacks inserted: {total_attacks}")
```

### Re-fetching with Validation

First pass was a bit sloppy — inserted attacks without checking if the war record was actually clean. Some wars had `clan_destruction > 100` because of an API quirk. This run checks that the war already exists in the DB before inserting attacks, so no orphaned or broken records sneak in.
